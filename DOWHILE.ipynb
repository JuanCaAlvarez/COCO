```python
# 1. Inicializa variables
suma_total = 0.0
cantidad_numeros = 0

print(" Calculadora de Promedio ")
print("Ingresa los números que deseas promediar.")

# 2. Utiliza un ciclo 'while True' para emular el Do-While
# El ciclo se ejecutará al menos una vez.
while True:

    # --- (Inicia el bloque "Do") ---

    # 3. Pide al usuario que ingrese un número
    try:
        entrada = input("Ingresa un número: ")
        numero = float(entrada)

        # 4. Suma su valor a la suma total y cuenta
        suma_total += numero
        cantidad_numeros += 1

    except ValueError:
        print("**Error: Por favor, ingresa un número válido.**")
        # Si la entrada es inválida, no continuamos y volvemos a preguntar
        continue

    # 5. Pregúntale al usuario si desea ingresar otro número
    respuesta = ""
    # Ciclo pequeño para validar la respuesta (s/n)
    while respuesta not in ['s', 'n']:
        respuesta = input("¿Deseas ingresar otro número? (s/n): ").lower().strip()

        if respuesta not in ['s', 'n']:
            print("Respuesta no válida. Escribe 's' para sí o 'n' para no.")

    # --- (Termina el bloque "Do" y verifica la condición "While") ---

    # 6. Verifica la respuesta al final del ciclo
    if respuesta == 'n':
        break # Si la respuesta es 'no', se rompe el ciclo principal

# 7. Si el usuario decide terminar, calcula el promedio
print("Procesando")

# 8. Verifica si se ingresó al menos un número (para evitar división por cero)
if cantidad_numeros == 0:
    print("**Error: No se ingresaron números.**")
    print("No se puede calcular el promedio.")
else:
    # Calcula y muestra el resultado
    promedio = suma_total / cantidad_numeros
    print(f"Cantidad de números ingresados: {cantidad_numeros}")
    print(f"Suma total: {suma_total}")
    print(f"El promedio es: -->  {promedio:.2f}  <---")

print("Gracias.")
```
