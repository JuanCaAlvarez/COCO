```python
# Inicializa la variable para acumular la suma
suma_pares = 0

print(" Suma de Números Pares del 1 al 100 ")


# GENERAR NUMEROS DEL 1 AL 100
for numero in range(1, 101):

    # VERIFICAR SI EL NUMERO ES PAR

    if numero % 2 == 0:

        # SUMA DE NUMEROS PARES
        suma_pares = suma_pares + numero


# SUMA TOTAL

print(f"La suma total de los números pares entre 1 y 100 es: -->  {suma_pares}  <--")
```
